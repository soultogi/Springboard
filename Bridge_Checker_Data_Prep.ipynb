```python
# Missing data check

def missingchecker(df):
    """ This function checks the missing value counts in every column and reports the number
    of missing values as well as the variable name, variables with no missing values will not be reported
    if there is no missing value in the whole data frame, no missing message will be reported """
    k=0
    for i in df.columns:
        j = sum(df[i].isnull())
        if j>0 :
            print('Variable Name : ', i,'  ---  Missing Counts : ', j)
            k =+ 1
    if k == 0:
        print('Unicorns are real :) , There are no missing values in the data frame')

missingchecker(df)
```
