```python
# Expanding Nested Data Batches with onl relevant columns

df_2 = json_normalize(data_raw, 'mjtheme_namecode', [ 'countryshortname', 'mjtheme'], errors='ignore')

print(df_2.head(3))

# more tidy dataframe indexing by 'countryshortname'

df_2 = df_2.set_index('countryshortname').sort_index()

print(df_2.head(3))

print(df_2.shape)
print(df_2.info())
print(df_2.columns)
```

```python
# QUESTION 2

print('**** TOP 10 PROJECT PROPOSAL CODES AND NAMES /W MISSING VALUES ****')
print(' ')
print('------------------------' + 'RAW DATA FRAME WITH NON-NORMALIZED DICTIONARY INFORMATION' + '------------------------')
print(' ')
print(df_1.mjtheme_namecode.value_counts().head(10))
print(' ')
print('------------------------' + 'TOP 10 MAJOR PROJECT CODES' + '------------------------')
print(' ')
q2_1 = df_2.code.value_counts()
print(df_2.code.value_counts().head(10))
print(' ')
print('------------------------' + 'TOP 10 PROJECT NAMES /W MISSING' + '------------------------')
print(' ')
q2_2 = df_2.name.value_counts()
print(df_2.name.value_counts().head(10))
```

```python
# DEALING WITH MISSING DATA
df_3 = df_2

df_3.tail()
```

```python
# CHECKING IF THE MISSING VALUES DEFINED OR NEEDS REDIFINITION

df_3.name.tail().isna()
```

```python
# REDEFINING MISSING VALUES

df_3 = df_3.replace('', np.nan)

df_3.tail()
```

```python
# CHECKING THAT EMPTY STRINGS ARE CODED AS Nan

print(df_3.name.tail().isna())

print(' ' )
#HOW MANY MISSING VALUES WE HAVE IN THE COLUMNS

print(df_3.info())
```

```python
## IN ORDER TO DEAL WITH THE MISSING VALUES I NEED TO USE CODE VALUES I HAVE EXTRACTED FROM THE MJTHEME_NAMECODE DICTIONARY
## BEFORE FILLING THE MISSINGS I NEED TO DOUBLE CHECK THAT EACH CODE LINKED WITH ONE UNIQUE NAME
## THEN I CAN FILL THE MISSING NAME VALUES WITH CODE MATCHES

# 'code' column read as STR in order to use it with logical filtering I transformed it into numeric values, 
# so more filtering tweaks can be applied.
df_3.code = pd.to_numeric(df_3.code)

#getting non-null values to create code based dictionary


code_list = df_3[df_3.name.notna()]

code_list.code.unique()

```

```python
code_dict={}

for i in code_list.code.unique():
    code_dict[i] = code_list[code_list.code == i].name.unique()
    
    
code_dict
```

```python
df_3 = df_3.reset_index()

print(df_3.tail())
```

```python
# since each code only relates with the one name, I can use it to deal with the missing values.

keyy = df_3[df_3.name.str.len() > 0].groupby(['code']).first().reset_index()

keyy = keyy[['code', 'name']]

keyy

#q3 = pd.merge(df_3[['code']], keyy, how='left')

#q3
```

```python
#QUESTION 3
# KEYY LIST I HAVE CREATED, WILL FILL THE MISSING DATA ON DF_3 BY MERGING IT ON KEY COLUMNS.

q3 = pd.merge(df_3, keyy, on='code', how='left')

q3 = q3.set_index('countryshortname').sort_index()[['code','name_y']]

q3.tail()
```

```python
#CHECKING THAT MISSINGS ARE FILLED

q3.info()
```

```python
#Getting Real Asnwers with unnested, missing cleaned data
print('-----------------------------' + 'TOP 10 COUNTRIES PROPOSED PROJECTS (CLEANED)' + '------------------------------')
print(q3.index.value_counts().head(10))
print(' ')
print('-----------------------------' + 'TOP 10 PROJECT CODES (CLEANED)' + '----------------------------')
print(q3.code.value_counts().head(10))
print(' ')
print('-----------------------------' + 'TOP 10 PROJECT NAMES (CLEANED)' + '----------------------------')
print(q3.name_y.value_counts().head(10))
```
